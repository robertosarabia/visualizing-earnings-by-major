## Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women')
```

## Findings
The more popular the major, the lower the median income. The percentage of women in any particular major has no correlation to median income although men do make, on average, more. Finally, majors with a higher percentages of full-time employed students also happen to have a slightly higher median income, on average. However, there is quite a lot of variance observed.

## Pandas, Histograms

```python
recent_grads['Sample_size'].hist(bins=20, range=(0,5000))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=40)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

## Findings
Women are somewhere between 60 to 70 percent across all majors. 30,000 to 40,000 is the most frequently recorded median income for recent majors.

## Pandas, Scatter Matrix Plot

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['ShareWomen', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

## Findings
There exists an inverse correlation between the popularity of a major and median income. Students who majored in subjects that were skewed in favor of female classmates made, on average, less. Recent majors made an average salary of 40K. There are on average between 60% and 70% females vs. male recent graduates.
